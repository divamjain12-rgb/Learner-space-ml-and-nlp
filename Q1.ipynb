```python
# Run this cell to install required libraries and import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Seaborn makes matplotlib plots look prettier!

print("Libraries imported successfully!")
```

### Step 1: Load the Data
Use `pandas` to read the CSV file from the provided URL.

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Load dataset
df = pd.read_csv(url)

# --- SANITY CHECK ---
assert 'df' in locals(), "Error: You didn't name your dataframe 'df'."
assert len(df) == 891, f"Expected 891 rows, got {len(df)}"
print("✅ Data loaded perfectly! Move to the next step.")


df.head()
```

### Step 2: Clean the Data
Real-world data is messy. Let's find out what is missing and fix it.

```python
# Print null values
print(df.isnull().sum())

# Fill missing Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

print("Missing values handled!")

# --- SANITY CHECK ---
assert df['Age'].isnull().sum() == 0, "Age column still has missing values!"
print("✅ Missing ages handled perfectly!")
df.head()
```

### Step 3: Visualize
Time to draw some graphs. Create at least 3 plots below.
Ideas: Survival rate by Gender, Passenger Class vs Survival, or an Age distribution histogram.

```python
# Create 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df, ax=axes[0])
axes[0].set_title('Survival by Gender')

# Plot 2: Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df, ax=axes[1])
axes[1].set_title('Survival by Class')

# Plot 3: Age Distribution
axes[2].hist(df['Age'], bins=30)
axes[2].set_title('Age Distribution')
axes[2].set_xlabel('Age')

# Adjust layout
plt.tight_layout()
plt.show()
```
